```python
import numpy as np
```

```python
data = np.genfromtxt("design_matrix.txt")
```

```python
print(data.shape)
nnumber = data.shape[0]*data.shape[1]
print(nnumber)
```

```python
missing_data = np.random.choice(nnumber, 5000)
```

```python
for i in missing_data:
    row = int(i/data.shape[0])
    column = i%data.shape[0]
    print(row,column)
    data[column, row] = 0
```

```python
np.savetxt("design_matrix-missing-data.txt",data)
```
